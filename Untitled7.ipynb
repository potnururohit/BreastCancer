```python
X=data_frame.drop("label",axis=1)
Y=data_frame['label']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
X.shape
```

```python
## standardize the data
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train_std=scaler.fit_transform(X_train)
X_test_std=scaler.transform(X_test)
```

```python
X_train_std
```

```python
Y.shape #relu means rectified linear units
```

## Building Neural Network

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

# no of output neurons=no of label classes
# since there are 2 classes 0 and 1 so two neurons

```python
# setting up the layers
model=keras.Sequential([keras.layers.Flatten(input_shape=(30,)),keras.layers.Dense(20,activation='relu'),keras.layers.Dense(2,activation='sigmoid')])
```

## if dog=0,cat=1,horse=2 ..if this type then we will use sparse_categorical_crossentropy
## if we have data of onehotencoding type then we will use categorical_crossentropy

```python
## compiling neural network
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
## training the neural network
history=model.fit(X_train_std,Y_train,validation_split=0.1,epochs=10)
# epochs how many times model has to go through the data
```

## loss function is nothing but the difference between the true value and predicted value
## So for any model the loss function must be minimum
## as loss value decreases the accuracy will increase

### Visualizing accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.xlabel('epoch')
plt.xlabel('accuracy')
plt.title('model accuracy')
plt.legend(['training data','validation_data'],loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.xlabel('epoch')
plt.ylabel('loss')
plt.title('model loss')
plt.legend(['training data','validation data'],loc='upper right')
```

## Accuracy of the model on test data

```python
loss,accuracy=model.evaluate(X_test_std,Y_test)
```

```python
input_data=(17.99,10.38,122.8,1001,0.1184,0.2776,0.3001,0.1471,0.2419,0.07871,1.095,0.9053,8.589,153.4,0.006399,0.04904,0.05373,0.01587,0.03003,0.006193,25.38,17.33,184.6,2019,0.1622,0.6656,0.7119,0.2654,0.4601,0.1189)
#change input data as numpy array
input_data_as_numpy_array=np.asarray(input_data)
#reshape the array
input_data_reshaped=input_data_as_numpy_array.reshape(1,-1)
#Standardizing data
input_data_std=scaler.transform(input_data_reshaped)
prediction=model.predict(input_data_std)
print(prediction)
prediction_label=[np.argmax(prediction)]
print(prediction_label)
if(prediction_label[0]==0):
  print('the tumor is Malignant so Breast Cancer')
else:
  print('the tumor is Bengin so no Breast Cancer')
```
